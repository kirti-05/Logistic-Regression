```python
train=training_data
train=train.drop('Name', axis=1)
train=train.drop('Ticket', axis=1)
```

```python
train.loc[train['Sex']=='female', 'Sex']=0
train.loc[train['Sex']=='male', 'Sex']=1
train.loc[train['Embarked'].isnull(), 'Embarked']='S'
train.loc[train['Embarked']=='S', 'Embarked']=1
train.loc[train['Embarked']=='C', 'Embarked']=2
train.loc[train['Embarked']=='Q', 'Embarked']=3
train.loc[train['Cabin'].isnull(), 'Cabin']=0
train.loc[train['Cabin']!=0, 'Cabin']=1
train.loc[train['Age'].isnull(), 'Age']=train.loc[train['Age'].notnull() , 'Age'].mean()
```

```python
train_data=train.values
```
